```python
from faker import Faker
import pandas as pd
import numpy as np
import faker 
class FakeDataGenerator:
    
    def __init__(self, df = None, rows=5):
        """
        This class generates fake data
        
        If data frame is passed in it will add the fake data to it. 
        Otherwise, it will create a new data frame
        
        Parameters:
        -----------
        df: 
          DataFrame default is None
          
        rows:
          If df None, this parameter will be used to create a data frame with `row` lenght
        """
        # code goes under this line. 
        # I will deduct points if any code is written above the comments!    

        self.profile = Faker()
        
        if df is not None:
            self.df = df
            self.rows = len(df)
        else:
            self.df = pd.DataFrame()
            self.rows = rows
        
        
    def add_random_number(self, column_name, small_value, high_value): # optional: add probabilities 
        """
        adds fake numbers to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        for i in range(1,25):
                  self.df[column_name]=[self.profile.random_int() for i in range(25)]
        return self.df       
    
        
    def add_first_name(self, column_name):
        """
        adds fake first name to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        for i in range(1,25):
                  self.df[column_name]=[self.profile.first_name() for i in range(25)]
        return self.df
    
    
    def add_last_name(self, column_name):
        """
        adds fake last name to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        for i in range(1,25):
                  self.df[column_name]=[self.profile.last_name() for i in range(25)]
        return self.df
              
    
    def add_email(self, column_name):
        """
        adds fake email to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        for i in range(1,25):
                  self.df[column_name]=[self.profile.email() for i in range(25)]          
        return self.df
    
    
    def add_phone_number(self, column_name):
        """
        adds fake name to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        for i in range(1,25):
                  self.df[column_name]=[self.profile.phone_number() for i in range(25)]
        return self.df
    
    
    def add_address(self, column_name):
        """
        adds fake name to the data frame with `column_name`
        
        Returns
        -------
        data frame
        """
        for i in range(1,25):
                  self.df[column_name]=[self.profile.address() for i in range(25)]
        return self.df
```
