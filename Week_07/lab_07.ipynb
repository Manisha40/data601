```python
import pandas as pd

url = 'https://raw.githubusercontent.com/mguner/teaching/main/datasets/nyc_flights/nyc_flights.csv'

flights = pd.read_csv(url)
```

```python
flights
```

```python
departuring_min = flights.groupby(['carrier','day']).agg({'dep_time' : 'min'})
  
departuring_min
```

```python
departuring_max = flights.groupby(['carrier','day']).agg({'dep_time' : 'max'})
  
departuring_max
```

```python
#When there is a missing value in dep_time then this is an indication of canceled flight. Find the number of cancelled flights for each (carrier, origin) combination.
flights.groupby('day').dep_time.min()
```

```python
#Which carrier has the worst dep_delays?
flights['dep_time'].isnull().value_counts()
```

```python
cancelled= flights.groupby(['origin','carrier']).agg({'dep_time':'count'})
cancelled
```

```python
worst = flights.groupby(["carrier"]).agg({'dep_delay':'max'})
worst
```

```python
#Which plane (tailnum) has the worst on-time record?

worst_record = flights.groupby(['tailnum']).agg(max_delay = ('dep_delay', 'max'))

max_value = worst_record['max_delay'].max()

worst_record[worst_record['max_delay'] == max_value]
```

```python
#For each plane, count the number of flights before the first delay of greater than 1 hour.

flights_count = flights.groupby(['tailnum', 'dep_delay']).agg(Flight_Count = ('flight', 'count'),
                                                              Departure_delay = ('dep_delay', 'max'),
                                                              Tail_Num = ('tailnum', 'max'))


flights_count
```

```python
flights_count[flights_count['Departure_delay'] > 1]
flights_count.groupby([flights_count['Tail_Num']]).agg({'Flight_Count' : 'sum'})
```
