```python
con = sqlite3.connect('data/flights.db')
cursor = con.cursor()
```

```python
cursor.execute("SELECT * FROM airports").fetchone()
```

```python
# Write a query that will join the latitude and longitude data from the airports table to the information on the routes table
cursor.execute("""
SELECT routes.*,latitude,longitude
FROM routes
LEFT JOIN airports
ON airports.code = routes.dest; """)
print(*cursor.fetchall(), sep='\n')
```

```python
# Which countries have the highest amount of active airports?
active_airports = cursor.execute("""
SELECT country,COUNT(active)
FROM airlines
WHERE active='Y'
GROUP BY (country) 
ORDER BY COUNT(active) DESC LIMIT 10;""")
print(*cursor.fetchall(), sep='\n')
```

```python
# Which countries have the highest amount of inactive airports?
cursor.execute("""
SELECT country, COUNT(active) AS inactive_airports
FROM airlines
WHERE active='N'
GROUP BY country
ORDER BY COUNT(active) DESC LIMIT 10;
""")
print(*cursor.fetchall(), sep='\n')
```

```python
# What about airports by timezones?
cursor.execute("""
SELECT timezone,count(country)
FROM airports
GROUP BY timezone
ORDER BY count(country) desc LIMIT 10;
""")
print(*cursor.fetchall(), sep='\n')
```

```python
# It's always a good idea to close our connections when we're done
cursor.close()
con.close()
```

# Integration

```python
conn = sqlite3.connect('data/Chinook_Sqlite.sqlite')
cur = conn.cursor()
```

```python
cur.execute(""" SELECT name FROM sqlite_master WHERE type='table';""")
print(cur.fetchall())
```

```python
cur.execute(""" SELECT * FROM sqlite_master WHERE type='table';""")
print(cur.fetchall())
```

```python
# What's the primary key of each table?
rows = cur.execute("SELECT name FROM sqlite_master WHERE type = 'table';")
tables = [row[0] for row in rows]
```

```python
def primary_key(s):
    return '"' + s.replace('"', '""') + '"'

for table in tables:
    print("table: " + table)
    rows = cur.execute("PRAGMA table_info({})".format(primary_key(table)))
    print(rows.fetchall())
```

```python
# What foreign keys join the tables together?
# hint: use "PRAGMA foreign_key_list()"
rows1 = cur.execute("SELECT name FROM sqlite_master WHERE type = 'table';")
tables1 = [row[0] for row in rows1]

def foreign_key(f):
    return '"' + f.replace('"', '""') + '"'

for table1 in tables1:
    print("table: " + table1)
    rows1 = cur.execute("PRAGMA foreign_key_list({})".format(foreign_key(table1)))
    print(rows1.fetchall())
```

```python
# What are the max and min dates in the Invoice table?
sql = '''
SELECT MAX (InvoiceDate) AS "Max Date", MIN (InvoiceDate) AS "Min Date"
FROM Invoice;
'''
pd.read_sql_query(sql, conn)
```

```python
#1. What are the names of all of the students.
sql = '''
SELECT name FROM responses;
'''
pd.read_sql_query(sql, conn)
```

```python
#2.Which student has the most siblings?
sql = '''
SELECT MAX(siblings_count), name 
FROM responses;'''
pd.read_sql_query(sql, conn)
```

```python
#3. How many students are only children? 
sql = '''
SELECT name,siblings_count 
FROM responses
WHERE siblings_count <1;
'''
pd.read_sql_query(sql, conn)
```

```python
#4. Which 3 students have lived in NYC the shortest amount of time? (How long has each lived in NYC?)
sql = '''
SELECT name, hometown,time_in_dc
FROM responses
WHERE hometown LIKE '%NY';
'''
pd.read_sql_query(sql, conn)
```

```python
#5.How many students are native New Yorkers?
sql = '''
SELECT name, hometown
FROM responses
WHERE hometown LIKE '%NY';
'''
pd.read_sql_query(sql, conn)
```

```python
#6. Do any two students have the same favorite food?
sql = '''
SELECT name, fav_food_1,fav_food_2
FROM responses
GROUP BY 1
HAVING fav_food_1 LIKE '%zza';
'''
pd.read_sql_query(sql, conn)

```

```python
# What are the favorite foods of this classroom?
sql = '''
SELECT name, fav_food_1,fav_food_2
FROM responses
'''
pd.read_sql_query(sql, conn)
```

```python
# Which student was born closest to the cohort's graduation date (5/29/20)?
sql = '''
select name,birthday from responses Where birthday >='5/29';  
'''
pd.read_sql_query(sql, conn)
```
