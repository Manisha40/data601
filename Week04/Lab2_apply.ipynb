```python
import pandas as pd
data_url="https://raw.githubusercontent.com/msaricaumbc/DS601_Fall21/main/Week04/data/bowling_stats.csv"
df=pd.read_csv(data_url,header=None)
```

```python
df
```

```python
df.columns =[ 'Year', 'City', 'State','N1','N2']
df
```

```python
#df["C"] = df[["City", "State"]].apply(lambda x: '{}{}'.format(x[0],x[1]), axis = 1)
df['c'] = df[["City","State"]].apply(lambda row: ','.join(row.values.astype(str)), axis=1)
```

```python
df.drop(['City', 'State'], axis = 1,inplace=True)
```

```python
df['Scored_points']  = df.apply(lambda x: x.N1*1000+x.N2, axis = 1)
```

```python
df.tail(40)
```

```python
df.Scored_points.fillna(df.N1, inplace=True)
df.drop(['N1', 'N2'], axis = 1,inplace=True)
```

```python
df.tail(40)
```

```python
df.plot.line(x = 'Year', y ='Scored_points', figsize =(12, 3), lw = 1)
```
