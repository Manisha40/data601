```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv("C://Users//Manisha//Downloads//RollingSystemDemand_20210928_2219.csv",
               index_col=False,skiprows=1,skipfooter=1,engine='python',header=None)

a.columns=['VD','Date_Time','Demand']
```

```python
a
```

```python
a['Date_Time']=pd.to_datetime(a['Date_Time'].astype(str),format='%Y%m%d%H%M%S')
a.head()
```

```python
a_2=pd.read_csv('C://Users//Manisha//Downloads//RollingSystemDemand_20210928_2221.csv',
                   index_col=False,skiprows=1,skipfooter=1,engine='python',header=None)
a_2.columns=['VD','Date_Time','Demand']
```

```python
a_2['Date_Time']=pd.to_datetime(a_2['Date_Time'].astype(str),format='%Y%m%d%H%M%S')
a_2.head()
```

```python
final_data=pd.concat([a,a_2])
final_data
```

```python
final_data.index=range(len(final_data))
```

```python
def hour_fun(x):
    return x.hour
final_data['tm_hour'] = final_data['Date_Time'].apply(hour_fun)
final_data['tm_date'] = pd.DatetimeIndex(final_data['Date_Time']).day
```

```python
final_data['tm_hour']
```

```python
final_data['tm_date']
```

```python
fig_size = plt.rcParams["figure.figsize"]
print(fig_size)
fig_size[0] = 12
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size
final_data['Demand'].groupby(final_data['tm_hour']).mean().plot(kind='bar', color='red')
plt.xlabel('Number of Hours', fontsize=20)
plt.ylabel('Average of demand in Mega Watts', fontsize=20)
plt.title('Mean value across all days per every hour', fontsize=20)
```

```python
final_data['Demand'].groupby([final_data['tm_date'],final_data['tm_hour']]).describe()
```

```python
fig_size = plt.rcParams["figure.figsize"]
print(fig_size)
fig_size[0] = 20
fig_size[1] = 150
plt.rcParams["figure.figsize"] = fig_size
final_data['Demand'].groupby([final_data['tm_date'],final_data['tm_hour']]).mean().plot(kind='barh',color='red')
plt.xlabel('Average Power Consumption in Mega Watts',fontsize=20)
plt.ylabel('DATE&HOUR',fontsize=20)
plt.title('Average Power Consumption per each hour on every day',fontsize=20)
```
